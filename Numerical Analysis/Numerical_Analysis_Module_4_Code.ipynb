## Problem 2

```python
def normMatrix(m):
    return(np.random.normal(loc=0, scale=np.sqrt(m), size = (m, m)))
```

(ii)(iii)

```python
eigVals = []
for i in range(1000):
    for m in [8,16,32,64,128,256,512,1024,2048, 4096]:
        eigVal, v = np.linalg.eig(normMatrix(8))
        eigVals.append(eigVal)

eigVals = np.array(eigVals).flatten()

fig = plt.figure()
plt.scatter(eigVals.real, eigVals.imag);
fig.suptitle('Eigenvalues of 1000 random normally distributed Matricies of size 8x8 to 4096x4096')

print np.max(np.absolute(eigVals))/np.min(np.absolute(eigVals.real))
```

## Problem 3
(a)

```python
def wilkin(n):
    w = np.add(np.diag([1.]*n) , np.tril(np.full((n,n), -1.), k=-1))
    w[:,-1] = 1.
    return(w)
```

(b)

(i)(ii)(iii)(iv)

```python
n = 60
W = wilkin(n)
e = np.ones(n)
b = np.dot(W,e)
x = np.linalg.solve(W,b)
print x
print e
```
